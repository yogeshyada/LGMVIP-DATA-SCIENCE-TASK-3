```python
target = data['targtype1_txt'].value_counts().head(10)
fig,ax = plt.subplots(figsize=(16,6))
ax.bar(target.index,target.values,color='b')
plt.title('Type of Targets')
plt.xlabel('Target')
plt.ylabel('Values')
plt.show()
```

So we can see the Private Citizens & Property were targeted more

```python
weapon_type = data['weaptype1_txt'].value_counts().head(5)
fig,ax = plt.subplots(figsize=(16,6))
ax.bar(weapon_type.index,weapon_type.values)
plt.title('Type of Weapons')
plt.xlabel('Weapons')
plt.ylabel('Values')
plt.show()
```

Explosives weapons are used more in attacks

```python
sns.displot(data,x='iyear',kind='kde',palette='dark',color='red')
sns.set_style('darkgrid')
```

```python
sns.boxplot(x=data['iyear'])
```

```python
cities_with_most_terrorism = data.city.value_counts().head(10)
cities = list(cities_with_most_terrorism.index)
cities
```

```python
fig,ax = plt.subplots(figsize=(16,6))
ax.bar(cities_with_most_terrorism.index,cities_with_most_terrorism.values)
plt.title('Cities with most terrorism')
plt.xlabel('City')
plt.ylabel('Value')
plt.show()
```

Here the Baghdad city has more number of terrorism attack

```python
sns.jointplot(x=data['iyear'],y=data['country'])
plt.figure(figsize=(10,10))
plt.style.use('ggplot')
```

```python
sns.distplot(a=data['country'],color='b')
```

```python
year = data['iyear'].unique()
years_count = data['iyear'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(16,6))
sns.barplot(x=year,y=years_count,palette='bright')
plt.xticks(rotation=80)
plt.xlabel('Attacking Year',fontsize=15)
plt.ylabel('No. of Attacks/Year',fontsize=15)
plt.title('Attacks in Years',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(data['iyear'],order=data['iyear'].value_counts().index,palette='bright')
plt.xticks(rotation=80)
plt.xlabel('AttackType',fontsize=15)
plt.ylabel('Counts',fontsize=15)
plt.title('Attacks/Year',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
data['country'].value_counts().to_frame().head(10).plot(kind='bar',color='g')
plt.xlabel('Terrorist group count',fontsize=15)
plt.ylabel('Attack number',fontsize=15)
plt.title('Top 10 terrorist group attack',fontsize=15)
plt.show()
```
